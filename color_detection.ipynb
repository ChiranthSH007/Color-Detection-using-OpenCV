```python
import numpy as np
import pandas as pd 
import cv2
```

```python
img = cv2.imread("colour.jpeg")
```

```python
index = ["color","color_name","hex","R","G","B"]
color_csv = pd.read_csv("colors.csv",names=index,header=None)
```

```python
clicked = False
r = g = b = xpos = ypos = 0
```

```python
def rec_col(R,G,B):
    minimum = 10000
    for i in range(len(color_csv)):
        d=abs(R-int(color_csv.loc[i,"R"])) + abs(G-int(color_csv.loc[i,"G"])) +abs(B-int(color_csv.loc[i,"B"]))
        print(d)
        if(d==minimum):
            minimum = d
            cname = color_csv.loc[i,"color_name"]
    return cname
```

```python
def mouse(event , x, y, flags, param):
    if(event == cv2.EVENT_LBUTTONDBLCLK):
        global b,g,r,xpos,ypos,clicked
        clicked = True
        xpos = x
        ypos = y
        b,g,r = img[y,x]
        b= int(b)
        g= int(g)
        r = int(r) 
```

```python
cv2.namedWindow("Colors Recognition")
cv2.setMouseCallback("Colors Recognition",mouse) 
while(1):
    cv2.imshow("Colors Recognition",img)

    if(clicked):
        cv2.rectangle(img,(20,20),(750,60),(b,g,r),-1)
        text=rec_col(r,g,b) + ' R='+str(r)+ ' G='+str(g)+ ' B='+str(b)
        print(text)
        cv2.putText(img, text,(50,50),2,0.8,(255,255,255),2,cv2.LINE_AA)
        if(r+g+b>=600):
            cv2.putText(img, text,(50,50),2,0.8,(0,0,0),2,cv2.LINE_AA)
        clicked=False
    if cv2.waitKey(20) & 0xFF ==27:
        break
cv2.destroyAllWindows()
```
